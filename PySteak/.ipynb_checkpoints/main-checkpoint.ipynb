```python
import csv
```

```python
# First, we need to define the path of the csv files for both menu items and sales data.
# On top of this, we need to define Menu and Sales lists to hold the contents of those respective CSV files.
# We're also using a loop to add the data from the files to the empty lists we just created.
menudata = Path("../Pysteak/menu_data.csv")
menu = []
with open(menudata, "r") as file:
    ramen_data = csv.reader(file)
    for row in file:
        csv_header = next(ramen_data)
        for row in ramen_data:
            menu.append(row)
        
```

```python
salesdata = Path("../Pysteak/sales_data.csv")
sales = []
with open(salesdata, "r") as file:
    sales_data = csv.reader(file)
    for row in file:
        csv_header = next(sales_data)
        for row in sales_data:
            sales.append(row)
```

```python
# Here, we're creating an empty dict to hold the data of menu and sales under several keys: Count, Revenue, Cogs, and Profit.
# We're useing for loops to define what rows contain what data, in order to calculate what will go into the keys
report = {}
for row in sales:
    quantity = int(row[3])
    sales_item = row[4]
    if sales_item not in report.keys():
        report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0,}
    for record in menu:
        item = record[0]
        price = float(record[3])
        cost = float(record[4])
        profit = price - cost
        if sales_item == item:
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += profit * quantity
        else: print(f"{sales_item} does not equal {item}! NO MATCH!")
```

```python
#Using a with loop, we're creating a new text file and writing the keys and values into  that text file to display our final analysis
with open("restaurant-analysis.txt", "w") as file:
    for key, value in report.items():
        file.write(f"{key}, {value}")
        
```
